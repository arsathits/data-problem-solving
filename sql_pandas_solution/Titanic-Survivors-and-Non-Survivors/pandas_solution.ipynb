```python
import pandas as pd
import numpy as np
import sys
import os

# Add project root to path
sys.path.append(r"D:\data-problem-solving")

from db import load_table, engine

titanic_df = pd.read_sql(
    'SELECT * FROM "titanic_survivor_non_survivor".titanic_data',
    engine
)
```

```python
result = (
    titanic_df.groupby("survived")
    .agg(
        first_class=("pclass", lambda x: (x==1).sum()),
        second_class=("pclass", lambda x: (x==2).sum()),
        third_class=("pclass", lambda x: (x==3).sum())                   
    )
    .reset_index()
)

print(result)
```
