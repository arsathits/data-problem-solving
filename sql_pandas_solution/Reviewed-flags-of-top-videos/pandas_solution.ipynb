```python
import pandas as pd
import numpy as np
import sys
import os

# Add project root to path
sys.path.append(r"D:\data-problem-solving")

from db import load_table, engine

# Load user_flags table
user_flags_df = pd.read_sql(
    'SELECT * FROM "reviewed_flags_top_videos".user_flags',
    engine
)

# Load flag_review table
flag_review_df = pd.read_sql(
    'SELECT * FROM "reviewed_flags_top_videos".flag_review',
    engine
)
```

```python
merged = pd.merge(user_flags_df, flag_review_df, on="flag_id", how= "inner")

flag_counts = merged.groupby("video_id").size().reset_index(name="total_flags")

max_flags = flag_counts["total_flags"].max()

top_videos = flag_counts[flag_counts["total_flags"] == max_flags]['video_id']

reviewed_counts = (
    merged[merged["video_id"].isin(top_videos) & merged['reviewed_by_yt'] == True]
    .groupby("video_id")
    .size()
    .reset_index(name='reviewed_flags')
)

print(reviewed_counts)
```
