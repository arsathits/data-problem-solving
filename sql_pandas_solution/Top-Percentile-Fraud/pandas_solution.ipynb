```python
import pandas as pd
import numpy as np
import sys
import os

# Add project root to path
sys.path.append(r"D:\data-problem-solving")

from db import load_table, engine

fraud_df = pd.read_sql(
    'SELECT * FROM "top_percentile_fraud".fraud_score',
    engine
)
```

```python
# Step 1. Compute 95th percentile per state
state_percentiles = (
    fraud_df.groupby('state')['fraud_score']
      .quantile(0.95)
      .reset_index()
      .rename(columns={'fraud_score': 'p95'})
)

# Step 2. Join back to the main dataframe
df_joined = fraud_df.merge(state_percentiles, on='state', how='left')

# Step 3. Filter for fraud_score >= p95
top_percentile_fraud = df_joined[df_joined['fraud_score'] >= df_joined['p95']]

# Step 4. Select required columns
result = top_percentile_fraud[['policy_num', 'state', 'claim_cost', 'fraud_score']]

print(result)
```
