```python
import pandas as pd
import numpy as np
import sys
import os

# Add project root to path
sys.path.append(r"D:\data-problem-solving")

from db import load_table, engine

df = pd.read_sql(
    'SELECT * FROM "popularity_percentage".facebook_friends',
    engine
)
```

```python
# Step 1: Create a bidirectional friendship table
users_union = pd.concat([
    df[['user1', 'user2']],
    df.rename(columns={'user2': 'user1', 'user1': 'user2'})[['user1', 'user2']]
])

# Step 2: Total number of unique users
total_users = users_union['user1'].nunique()

# Step 3: Count number of friends per user
popularity = (
    users_union.groupby('user1')
    .size()
    .reset_index(name='friend_count')
)

# Step 4: Calculate popularity percentage
popularity['popularity_percent'] = (popularity['friend_count'] / total_users) * 100

# Step 5: Order by user ID
popularity = popularity.sort_values('user1').reset_index(drop=True)

print(popularity)
```
